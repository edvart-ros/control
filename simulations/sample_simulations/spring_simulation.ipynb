```python
from scipy import signal as sig
import control
import numpy as np
import matplotlib.pyplot as plt
#Simulation of mass-spring-damper system, formulated in state space form and simulated with scipy and control

#Differential equation:
#mx'' = u - kx - cx'

# Parameters
m = 20
k = 2
c = 4

#simulation parameters
start = 0
end = 40
dt = 0.1
t = np.arange(start, end, dt)

# State-space matrices
A = np.array([[0, 1], [-k/m, -c/m]])
B = np.array([[0], [1/m]])
C = np.eye(2)  # only consider the first state variable
D = np.zeros((2, 1))

```

```python
sys = sig.StateSpace(A, B, C, D)
x0 = [1, 0]
u = np.ones(len(t))
T, y, x = sig.lsim(sys, u, t, X0=x0)

#plot the step response
#set ylim

plt.plot(t, y)
plt.grid(True)
```
